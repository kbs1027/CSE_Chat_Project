```python
import feedparser
import schedule
import time

parse_rss = feedparser.parse("https://cse.pusan.ac.kr/bbs/cse/12278/rssList.do?row=50")
```

```python
for p in parse_rss.entries:
    date = f'{p.published_parsed.tm_year}-{p.published_parsed.tm_mon}-{p.published_parsed.tm_mday}'
    print("경진대회 게시판\n")
    print(f'{date}\n{p.title}\n{p.link}')
```
